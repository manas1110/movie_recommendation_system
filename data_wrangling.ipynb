CONVERTING THE DURATION COLUMN INTO MINUTES

```python
def func(x):
    if ('H' in x) and ('M' in  x):
        a=int(x.split('H')[0])*60
        b=x.split('H')[1].split('M')[0].strip()
        if b=='':
            b=0
        else:
            b=int(x.split('H')[1].split('M')[0])
        return (a+b)
    elif 'H' in x:
        a=int(x.split('H')[0])*60
        return a
    else:
        b=int(x.split('M')[0])
        return b

df_1['duration']=df_1['duration'].apply(func)
```
